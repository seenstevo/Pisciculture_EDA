```python
aquaculture_quantity_value = pd.read_csv("./data/Aquaculture_Quantity_Code_Merged.csv", 
                                         usecols = ['Environment', 'PERIOD', 'Quantity', 'Country', 'WaterArea', 'Species', 'Scientific_Name', ])
```

```python
population_by_country = pd.read_csv("./data/Population_by_Country.csv")
population_by_country.head()
```

```python
aquaculture_quantity_value = pd.merge(aquaculture_quantity_value, population_by_country)
aquaculture_quantity_value['Population'] = aquaculture_quantity_value['Population'] / 1000000
aquaculture_quantity_value.head()
```

```python
aquaculture_quantity_value['Quantity_Population_Normed'] = aquaculture_quantity_value['Quantity']  / aquaculture_quantity_value['Population']
# tonnes produced per million people
aquaculture_quantity_value.drop(['Quantity', 'Population'], axis = 1, inplace = True)
aquaculture_quantity_value.head()
```
