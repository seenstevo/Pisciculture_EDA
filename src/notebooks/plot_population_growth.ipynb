```python
import seaborn as sns
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
os.getcwd()
```

```python
pop = pd.read_csv("../../misc/world_pops_continent.csv", index_col=0)
```

```python
pop = pop.T
```

```python
pop = pop.drop(columns = 'World ')
```

```python
pop = pop/1000000
```

```python
pop.plot(kind = 'bar', stacked = True)
plt.legend(bbox_to_anchor=(1, 1))
plt.title("Projected Population Growth This Century")
plt.ylabel("Population in Billions")
```
